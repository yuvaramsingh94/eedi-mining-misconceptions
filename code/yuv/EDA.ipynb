## Competition data

```python
train_df = pd.read_csv(os.path.join(DATASET, "competitions","eedi-mining-misconceptions-in-mathematics","train.csv"))
## COnvert column MisconceptionAId from float to string
train_df['MisconceptionAId'] = train_df['MisconceptionAId'].astype(str)
train_df['MisconceptionBId'] = train_df['MisconceptionBId'].astype(str)
train_df['MisconceptionCId'] = train_df['MisconceptionCId'].astype(str)
train_df['MisconceptionDId'] = train_df['MisconceptionDId'].astype(str)
train_df.head()
```

```python
## print the number of unique QuestionId, SubjectId, SubjectName
print("Unique QuestionId:", train_df['QuestionId'].nunique())
print("Unique SubjectId:", train_df['SubjectId'].nunique())
print("Unique SubjectName:", train_df['SubjectName'].nunique())
## Print the number of unique MisconceptionAId, MisconceptionBId, MisconceptionCId, MisconceptionDId. here we have to combine all the misconceptions into one and get the unique values
misconceptions = pd.concat([train_df['MisconceptionAId'], train_df['MisconceptionBId'], 
                            train_df['MisconceptionCId'], train_df['MisconceptionDId']])
print("Unique MisconceptionId:", misconceptions.nunique())
```
